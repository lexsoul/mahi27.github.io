```python
def listen_to_audio(path, category):   
    rate, wav_sample = wav.read(path)
    wav_file = open(path,"rb")
    print('Sampling rate: ',rate,'Hz')
    print('Duration: ',wav_sample.shape[0]/rate,' second')
    print('No.of samples: ',len(wav_sample))
    print('Category: ',category)
    plt.figure(figsize=(12, 4))
    plt.plot(wav_sample) 
    return ipd.Audio(path)
```
